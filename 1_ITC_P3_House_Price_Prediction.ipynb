# KNN Classification

```python
### Evaluate KNN with bathrooms,fireplaces,HOA,latitude,longitude as Independent features and
#### price per sqft category as Dependent feature
```

```python
# KNN Measures distance between 2 points called euclidean distance

class KNNClassifier():

  def fit(self, X, y):
    self.X = X
    self.y = y

  def predict(self, X, K, epsilon = 1e-3):    # K is number of neighbours
    N = len(X)                                # Total no of observations
    y_hat = np.zeros(N)                       # array of predictions of size N

    #get distances , loop through every point
    for i in range(N):
      dist_square = np.sum((self.X-X[i])**2, axis = 1)

      # Get the indexes of K lowest values of distances (idxt is index of nearest point)
      idxt = np.argsort(dist_square)[:K]
      gamma_k = 1 / (np.sqrt(dist_square[idxt] + epsilon))

      ##classes of nearest neighbours, gamma weights (inverse of distances)
      # bincount returns array of  weighted sum for each class
      y_hat[i] = np.bincount(self.y[idxt], weights= gamma_k).argmax()

    return y_hat
```

```python
def accuracy(y,y_hat):
  return np.mean(y==y_hat)
```

# KNN Regressor

```python
class KNNRegressor():

  def fit(self, X, y):
    self.X = X
    self.y = y

  def predict(self, X, K, epsilon = 1e-3):
    N = len(X)
    y_hat = np.zeros(N)

    # Loop through every point
    for i in range(N):
      dist_sq = np.sum((self.X - X[i])**2, axis = 1)                    # square of distances along dimensions
      idxt = np.argsort(dist_sq)[:K]                                    # distances sorted from lowest, K is number of lowest distances indexes
      gamma_k = np.exp(-dist_sq[idxt]) / np.exp(-dist_sq[idxt]).sum()
      y_hat[i] = gamma_k.dot(self.y[idxt])

    return y_hat
```

```python
def R2(Y, Y_hat):
  return (1-(np.sum((Y-Y_hat)**2)/np.sum((Y-np.mean(Y))**2)))
```

```python
import numpy as np
from geopy import Nominatim
import folium

def House_price_prediction_address():

    ##lat_coord,long_coord = get_coordinates(data_model)
    geolocator = Nominatim(user_agent = "IGT")

    ## Get the input address from user
    try:
       addr      = input("Please give the address to look for : ")
       location  = geolocator.geocode(addr)
       lat_pred  = location.latitude
       long_pred = location.longitude
    except:
       print("Address could not be encoded by geopy.")
       lat_pred,long_pred = get_lat_long_user_inputs()

    yclass = get_psf_category(lat_pred, long_pred)

    price,rate_pred = get_price(yclass, lat_pred, long_pred)
    map_loc = plot_map(price, rate_pred, lat_pred, long_pred)
    return map_loc


def get_lat_long_user_inputs():
    ## Get the latitude and longitude from user
    lat_pred_str  = input("Latitude : ")
    long_pred_str = input("Longitude : ")
    lat_pred  = float(lat_pred_str)
    long_pred = float(long_pred_str)
    return lat_pred,long_pred


def get_psf_category(lat_pred, long_pred):
    ## ---------------------------------------------------
    ## Step1 - KNN Classification
    ## Derive class category for given Latitude, longitude,
    ## X=[per_sqft, latitude, longitude], y=[ppsfqt_class]
    ##----------------------------------------------------
    features = ['ppsqft_class','latitude','longitude']
    X_y_data = data_model[features]
    X_y_data = X_y_data.to_numpy()
    ytrain = X_y_data[: , 0]
    ytrain = ytrain.astype('int64')
    Xtrain = X_y_data[: , 1:]

    knn = KNNClassifier()
    knn.fit(Xtrain, ytrain)

    Xarr = np.array([[lat_pred, long_pred]])
    yclass = knn.predict(Xarr, 10)
    return yclass


def get_price(yclass, lat_pred, long_pred):
    ## ---------------------------------------------------
    ## Step 2 - KNN Regressor
    ## Find the per_sqft rate for the class_category along with
    ## no. of bathrooms, no. of fireplaces, area of the house(sqrt_ft)
    ##----------------------------------------------------
    area_str      = input("Area of house in sqft : ")
    bathrooms_str = input("Number of bathrooms   : ")
    fireplaces_str= input("Number of fireplaces  : ")
    area = float(area_str)
    bathrooms = float(bathrooms_str)
    fireplaces = float(fireplaces_str)

    features = ['per_sqft','sqrt_ft', 'bathrooms', 'fireplaces', 'ppsqft_class']
    X_y_data = data_model[features]
    X_y_data = X_y_data.to_numpy()
    ytrain = X_y_data[: , 0]
    Xtrain = X_y_data[: , 1:]

    knnreg = KNNRegressor()
    knnreg.fit(Xtrain, ytrain)

    Xarr = np.array([[area, bathrooms, fireplaces, float(yclass)]])
    rate_pred = knnreg.predict(Xarr,10)
    price = area * rate_pred
    return price, rate_pred


def plot_map(price, rate_pred, lat_pred, long_pred):
    lat_long_loc = folium.Map(location=(lat_pred, long_pred), zoom_start = 20)

    (folium.Marker(location=(lat_pred, long_pred),
                  popup=f"Predicted price of the house is ${price} at {rate_pred} psf",
                  icon = folium.Icon(color = 'blue'))).add_to(lat_long_loc)

    return lat_long_loc
```
